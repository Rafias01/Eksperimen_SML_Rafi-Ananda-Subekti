# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
#Type your code here
train_dir = os.path.join(extract_path, "train")

classes = os.listdir(train_dir)
print("Kelas kendaraan:", classes)

image_count = {}

for cls in classes:
    cls_path = os.path.join(train_dir, cls)
    image_count[cls] = len(os.listdir(cls_path))

image_count
```

```python
plt.figure(figsize=(8,5))
plt.bar(image_count.keys(), image_count.values())
plt.title("Distribusi Jumlah Gambar per Kelas")
plt.xlabel("Kelas Kendaraan")
plt.ylabel("Jumlah Gambar")
plt.show()
```

```python
sample_class = classes[0]
sample_path = os.path.join(train_dir, sample_class)

sample_image = os.listdir(sample_path)[0]
img = Image.open(os.path.join(sample_path, sample_image))

plt.imshow(img)
plt.title(f"Contoh gambar kelas: {sample_class}")
plt.axis("off")
plt.show()
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
IMG_SIZE = (224, 224)

def load_and_preprocess_image(img_path):
    img = Image.open(img_path).convert("RGB")
    img = img.resize(IMG_SIZE)
    img = np.array(img) / 255.0
    return img
```

```python
X = []
y = []

for cls in classes:
    cls_path = os.path.join(train_dir, cls)
    for img_name in os.listdir(cls_path):
        img_path = os.path.join(cls_path, img_name)
        X.append(load_and_preprocess_image(img_path))
        y.append(cls)

X = np.array(X)
y = np.array(y)

print("Shape X:", X.shape)
print("Shape y:", y.shape)
```

```python
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

print("Mapping label:", dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Train shape:", X_train.shape)
print("Validation shape:", X_val.shape)
```
